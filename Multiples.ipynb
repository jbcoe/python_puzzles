```python
import numpy.random as r
```

```python
NUMBERS = [3,4,5,6,7,8,9,11,12]
```

```python
def division_problem():
    x = r.choice(NUMBERS)
    y = r.choice(NUMBERS)

    z = x * y
    
    pick = r.randint(3)
    
    if pick == 0:
        return f"{x:2} x {y:2} =   "
    if pick == 1:
        return f"{x:2} x    = {z:2}"
    if pick == 2:
        return f"   x {y:2} = {z:2}"
```

```python
division_problem()
```

```python
def generate_problems(*, count, seed):
    r.seed(seed)
    text = """Maths problems #{seed}
    
{problems}
""".format(seed=seed, problems = "\n\n".join(
        [ "{}      {}      {}      {}".format(
            division_problem(), 
            division_problem(), 
            division_problem(), 
            division_problem()) 
         for _ in range(count//4)]))
    
    print(text)
    with open(f"/tmp/maths_problems_{seed}.txt", "w") as o:
        o.write(text)
```

```python
for seed in range(1, 5):
    generate_problems(count=100, seed=seed)
```
